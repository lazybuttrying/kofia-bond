```python
def rounding(v):
    money, currency = v
    if not money:
        return None
    money = float(money)
    if currency == "억원":
        return money / 100000000
    elif currency == "백만달러":
        return money / 1000000
    elif currency == "백만EUR":
        return money / 1000000
    elif currency == "억엔":
        return money / 100000000
```
