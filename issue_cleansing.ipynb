Data type

```python
for c in kofia.columns:
    try:
        t = kofia[kofia[c].apply(lambda x: isinstance(x, str))]
        tt =  kofia[kofia[c].apply(lambda x: isinstance(x, float))]
        tt = tt[~tt[c].isna()]
        ttt = kofia[kofia[c].apply(lambda x: isinstance(x, int))]
        if t.shape[0] > 0 and (tt.shape[0] > 0 or ttt.shape[0] > 0):
            print("---", c)
        if t[c].str.contains(",").sum():
            print(c)
    except:
        pass
```

기간

```python
kofia['기간'] = (pd.to_datetime(kofia['만기일'].apply(lambda x: "2262-04-11" if x=="9999-12-31" else x)) - pd.to_datetime(kofia['발행일'])).dt.days
kofia.loc[kofia['만기일']=="9999-12-31", '기간'] = -1
```

```python
kofia.loc[kofia["이자지급 선/후급"].isna(), "이자지급 선/후급"] = "-"
```

원리금지급기관

```python
kofia.loc[(kofia["원리금지급기관"]=="000000000000"), "원리금지급기관"] = None
kofia.loc[kofia["원리금지급기관"].isna(), "원리금지급기관"] = None
```

주관회사

```python
kofia.loc[kofia["주관회사"]=="00000000000000000000", "주관회사"] = None
kofia.loc[kofia["주관회사"].isna(), "주관회사"] = None
```

선매출유형

```python
kofia.loc[(kofia["선매출유형"].isna()), "선매출유형"] = "-"
```

상장일, 선매출일

```python
kofia.loc[(kofia["상장일"].isna()) | (kofia["상장일"]=="0--"), "상장일"] = "0000-00-00"
kofia.loc[(kofia["선매출일"].isna()) | (kofia["선매출일"]=="0--"), "선매출일"] = "0000-00-00"
```

상장구분

```python
kofia.loc[kofia["상장구분"].isna(), "상장구분"] = "-"
```

종목명

```python
kofia["종목명"] = kofia["종목명"].str.strip()

for isin in ["KR2001014B24", "KR3520029E63", "KR6001741D44", "KRC035AP2562"]:
    kofia.loc[kofia["표준코드"] == isin, "종목명_detail"] = kofia.loc[kofia["표준코드"] == isin, "종목명"]

kofia["종목명"] = kofia["종목명_detail"]
del kofia["종목명_detail"]
```
